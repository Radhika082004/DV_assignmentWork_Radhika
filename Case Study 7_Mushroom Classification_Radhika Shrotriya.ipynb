```python
# Generating a dataset with multiple columns

np.random.seed(42)
data = pd.DataFrame(np.random.randn(100, 24), columns=[f'Column_{i}' for i in range(24)])

# Defining the number of columns in the dataset

num_columns = len(data.columns)

# Setting up subplots for multiple distribution plots

fig, axes = plt.subplots(nrows=num_columns, ncols=1, figsize=(8, 2*num_columns))
plt.subplots_adjust(hspace=0.5)

# Create distribution plots for each column using a for loop

for i, column in enumerate(data.columns):
    ax = axes[i]
    sns.histplot(data=data, x=column, kde=True,ax=ax)
    ax.set_xlabel(column)
    ax.set_ylabel('Count')

# Show the plots
plt.show()
```
